## What is the most commonly awarded gender and birth country?
### Storing answers as strings at top_gender and top_country

```python
df = pd.read_csv('data/nobel.csv')
df.head(10)
```

## Which decade had the highest ratio of US-born Nobel Prize winners to total winners in all categories?
### Stored as max_decade_usa

```python
# Add a 'decade' column
df['decade'] = (df['year'] // 10) * 10

# Filter for USA and group by 'decade', then count occurrences
decade_usa = df[df['birth_country'] == 'United States of America']
max_decade_usa = decade_usa['decade'].value_counts().idxmax()
print(max_decade_usa)
```

```python
top_gender = 'Male' if df['sex'].eq('Male').sum() > df['sex'].eq('Female').sum() else 'Female'
top_country = df['birth_country'].value_counts().idxmax()
print('top gender: ', top_gender)
print('top country: ', top_country)
```

## Which decade and Nobel Prize category combination had the highest proportion of female laureates?
### Storing this as a dictionary called max_female_dict

```python
# Filter for female laureates
female_df = df[df['sex'] == 'Female']

# Count total laureates per decade and category
total_counts = df.groupby(['decade', 'category']).size().rename('total_count')

# Count female laureates per decade and category
female_counts = female_df.groupby(['decade', 'category']).size().rename('female_count')

# Combine counts into a single DataFrame
counts_df = pd.concat([total_counts, female_counts], axis=1).fillna(0)

# Calculate the proportion of female laureates
counts_df['female_proportion'] = counts_df['female_count'] / counts_df['total_count']

# Find the row with the highest proportion of female laureates
max_row = counts_df.loc[counts_df['female_proportion'].idxmax()]

# Store as a dictionary: {decade: category}
max_index = counts_df['female_proportion'].idxmax()
max_female_dict = {int(max_index[0]): max_index[1]}
max_female_dict
```

## Who was the first woman to receive a Nobel Prize, and in what category?
### Saving strings as first_woman-name and first_woman_category

```python
# Corrected code
first_woman_name = female_df['full_name'].iloc[0]
first_woman_category = female_df['category'].iloc[0]
print('Name and Category: ', first_woman_name, 'and', first_woman_category)
```

## Which individuals or organizations have won more than one Nobel Prize throughout the years?
### Storing full names in a list named repeat_list

```python
# Find laureates (individuals or organizations) who have won more than one Nobel Prize
repeat_counts = df['full_name'].value_counts()
repeat_list = repeat_counts[repeat_counts > 1].index.tolist()
repeat_list
```
