Duration variable visualization and most frequent **MOVIE** from the **1990s**

```python
netflix_movie_df = netflix_df[(netflix_df['type'] == 'Movie') & (netflix_df['release_year'] >= 1990) & (netflix_df['release_year'] < 2000)]
```

```python
netflix_movie_df.head()
```

```python
plt.hist(netflix_movie_df['duration'])
plt.xlabel('Duration')
plt.ylabel('Frequency')
plt.title('Frequency of Movie Durations on Netflix')
plt.show()
```

As we can see, the most frequent duration is around 100 minutes.

Short movies (less than 90 minutes)

```python
short_movie_count = netflix_movie_df[(netflix_movie_df['duration'] < 90) & (netflix_movie_df['genre'] == 'Action')]
short_movie_count = short_movie_count.shape[0]
print(short_movie_count)
```
