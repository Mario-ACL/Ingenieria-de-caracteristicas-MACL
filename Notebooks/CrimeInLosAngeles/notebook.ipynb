Which hour has the highest frequency of crimes? The result will be saved as an int in peak_crime_hour

```python
print(crimes['TIME OCC'].str.slice(stop=2).value_counts())
peak_crime_hour = int(crimes['TIME OCC'].str.slice(stop=2).value_counts().idxmax())
```

Which area has the largest frequency of night crimes (crimes committed between 10pm and 3:59am)? The result will be saved as a string in peak_night_crime_location

```python
area_night_df = crimes[(crimes['TIME OCC'].astype(int) > 2200) | (crimes['TIME OCC'].astype(int) < 400)]
# Now we count
peak_night_crime_location = area_night_df['AREA NAME'].value_counts().idxmax()
print(peak_night_crime_location)
```

Identify the number of crimes committed against victims of different age groups. Will be saved as a pandas Series called victim_ages, with age group labels "0-17", "18-25", "26-34", "35-44", "45-54", "55-64", and "65+" as the index and the frequency of crimes as the values.

```python
age_groups = ["0-17", "18-25", "26-34", "35-44", "45-54", "55-64", "65+"]

age_group1 = crimes[crimes['Vict Age'] <= 17]['Vict Age'].value_counts().sum()
age_group2 = crimes[(crimes['Vict Age'] <= 25) & (crimes['Vict Age'] >= 18)]['Vict Age'].value_counts().sum()
age_group3 = crimes[(crimes['Vict Age'] >= 26) & (crimes['Vict Age'] <= 34)]['Vict Age'].value_counts().sum()
age_group4 = crimes[(crimes['Vict Age'] >= 35) & (crimes['Vict Age'] <= 44)]['Vict Age'].value_counts().sum()
age_group5 = crimes[(crimes['Vict Age'] >= 45) & (crimes['Vict Age'] <= 54)]['Vict Age'].value_counts().sum()
age_group6 = crimes[(crimes['Vict Age'] >= 55) & (crimes['Vict Age'] <= 64)]['Vict Age'].value_counts().sum()
age_group7 = crimes[(crimes['Vict Age'] >= 65)]['Vict Age'].value_counts().sum()

age_values = [age_group1, age_group2, age_group3, age_group4, age_group5, age_group6, age_group7]

victim_ages = pd.Series(age_values, index=age_groups)

victim_ages.head()
```
